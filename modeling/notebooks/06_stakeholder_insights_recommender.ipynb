# Stakeholder Insight Recommender (via PDP Analysis)

### Developers
**DAU Impact on Revenue**

**Insights:**
* Revenue per user (RPU) declines as DAU (Daily Active Users) increases,suggesting diminishing marginal returns from expanding the active user base alone.

**Action:**
* Focus should shift toward optimizing **user experience quality and engagement**, not just quantity.
* Developers may explore techniques that optimize backend infrastructure to support peak DAU periods more efficiently without overspending.

### Product Managers

1. **Monetization Metrics**

**Insight:** `purchase_per_session` - Average Purchase per Session vs Monetization
* There is a positive relationship between purchases per session and revenue per user.
* As purchases increase beyond ~1000 per session, the monetization impact starts to plateau.

**Action:** Optimize features or offers that encourage mid-session purchases witout pushing for saturation.


2. **Release Timing Trends**

**Insight:** `release_dayofweek_freq` - Day of Release Influence on Revenue
* Minor variation suggests that certain days of the week may offer slightly higher monetization outcomes.

**Action:** Experiment with targeted promotions or releases on days that historically show marginal gains.

**Insight:** `release_month_freq` - Seasonal Release Trends impact on Revenue
* Higher release activity in specific months appears correlated with lower revenue

**Action:** Spread releases across the calendar or revisit content strategy during high-volume months to maintain revenue levels.


3. **Player tier Performance**

**Insight:** 
* `dau_bin_Low` : Low DAU Tier has strong positive impact on monetization, mostly niche engaged user base.
* `dau_bin_Medium` :  Medium DAU Tier shows positive monetization correlation, with ideal balance of scale and enagagement.
* `dau_bin_High` and `dau_bin_Very High` : High and Very High DAU Tiers have negative impact on RPU, possibly due to diluted enagagement or freemium users.

**Action:** Focus monetization experiments on Low to Mid DAU segments and revisit enagagement or segmentation strategies for higher tiers.

### Marketers

1. **New Player Growth**

**Insight:**
* `scaled_new_registrations` shows dimishing returns on mentization.
* While initial user growth boosts revenue, the marginal impact decreases over time.

**Action:**
* Focus on improving conversion and retention instead of just acquiring users.

2. **Social Media Buzz**

**Insight:**
* For `scaled_social_media_mentions`, Revenue Per User peaks at moderate buzz levels and drops when mentions are too low or too high.
    * Suggesting oversaturation of hype fatigue.
 
**Action:**
* Calibrate marketing campaigns to maintain optimal enagagement levels rather than pushing excessive volume.

3. **Stream Viewership**

**Insight:**
* Overall `scaled_stream_viewership` shows a negatuve correlation with revenue, possibly due to passive viewers not converting into paying users.

**Action:**
* Shift strategies to engage viewers actively, converting attention into in-game actions or purchases.

4. **Stream Popularity per Player**

**Insight:**
* A higher stream-to-player (`stream_viewers_per_player`) ratio is strongly correlated with increased monetization.

**Actions:**
* Prioritize streamable moments, influencer partnerships and features that highlight individual gameplay.

5. **Influencer Endorsements**

**Insight:**
* `influencer_endorsements` linearly improve revenue outcomes, indicating consistent value from craetor partnerships.

**Action:**
* Continue or scale influencer collaboration to reach targeted, high-conversion audiences.

### Designers

1. **Session Duration & Engagament**

**Insight:**
* Longer session durations (`scaled_session_duration_minutes`) correlate with higher revenue per user.

**Action:**
* Suggests that designers should optimize game flow and reduce friction to encourage longer playtimes.


2. **Session Intentsity (DAU x Duration)**

**Insight:**
* Extremely high user session (`dau_x_duration`) volume does not proportionally increase revenue.
* There may be dinimishing returns or saturation.

**Action:**
* Need to balance game pacing and load handling.


3. **Enagagement Level**

**Insight:**
* Players with high enagament levels (`enagament_level_High`), drive more revenue.

**Action:**
* Designers should enhance features that deepen player involvement, such as personalized missions or dynamic content unlocks.

### Publishers

1. **In-Game Spend**

**Insight:**
* As in-game purchases (`scaled_ingame_purchases_`) increase, the predicted revenue per user (RPU) shows a steady and strong upward trend.

**Action:**
* This reinforces the value of in-game monetization strategies such as in-game currency packs and cosmetic upgrades.
* Encouraging in-game transactions through limited offers or rewarding enagagement loops could improve monetization.


2. **Platform Distribution**

**Insight**
* Higher frequency of game releases on certain platforms (`platform_freq`) is positively correlated with RPU.

**Action:**
* Strategic platform targeting is important.
* Optimizing launch efforts on platforms with historically higher monetization potential, such as PC or Console, could yield better financial outcomes.


3. **Popular Genres**

**Insights:**
* Genre frequency (`top_genre_freq`) shows a U-saped effect, where both low and high frequencies slightly improve monetization compared to the mid-range.

**Action:**
* Balancing genre saturation can be the key.
* Oversaturation of popular genres may stagnate returns, while niche genres or genre hybrid might provide new monetization opportunites.

### Conclusion

This capstone project explored how player behavior, platform trends and in-game features influence monetization outcomes across stakeholder groups in a gaming industry.

Using Machine Learning and Partial Dependence Analysis, key patterns were uncovered, such as the declining marginal returns of DAU (Daily Active Users) and new registrations, the monetization potential of high engagement segments, and the effects of Genre-Platform strategies.

**Random Forest with RandomizedSearchCV** emerged as the best=performing model (Test R2 ~63.34%),offering a robust foundation for stakeholder-specific insights.

Additionally, **the dataset was intentionally kept with outliers** to reflect real-world scenarios and **preserve signals from high-value users**, which are especially relevant in gaming monetization analysis.

The **stakeholder-aligned PDP recommender framework delivered actionable insights** tailored for product managers, marketers, designers, developers and publishers, **bridging predictive modeling with business relevance.**
